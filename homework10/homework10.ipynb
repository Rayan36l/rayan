```python
df = pd.read_csv("solar_system.csv")
df_original = df.copy()
df = df.set_index("Attribute").T
df.index.name = "Planet"
df.reset_index(inplace=True)
df.columns.name = None
```

```python
#2.1.3 
cols_to_skip = ["Planet", "Ring System?", "Global Magnetic Field?"]
for col in df.columns:
    if col not in cols_to_skip:
        df[col] = pd.to_numeric(df[col], errors='coerce')

print(df.dtypes)
```

```python
#2.2
def attach_units(col_name, unit, new_name=None):
    df[col_name] = [val * unit for val in df[col_name]]
    
    if new_name:
        df.rename(columns={col_name: new_name}, inplace=True)
```

```python
# Picking Gravity 
unit_grav = u.m / (u.s**2)
old_name = "Gravity (m/s^2)"
new_name = "Gravity (m/s2)"

attach_units(old_name, unit_grav, new_name)

print(df[new_name])
```

```python
#2.3 Data Analysis
# 1. Calculating semi-major axis
peri_vals = [v * 1e6 * u.km for v in df["Perihelion (10^6 km)"]]
aphe_vals = [v * 1e6 * u.km for v in df["Aphelion (10^6 km)"]]

sma_km = [(p + a) / 2 for p, a in zip(peri_vals, aphe_vals)]

df.insert(df.columns.get_loc("Aphelion (10^6 km)") + 1, "Semi-Major Axis (km)", sma_km)

# Conversion
df["Semi-Major Axis (AU)"] = [v.to(u.AU) for v in df["Semi-Major Axis (km)"]]


df.to_csv("units.csv", index=False)
print("Saved units.csv successfully!")


df[["Planet", "Semi-Major Axis (km)", "Semi-Major Axis (AU)"]].head()
```
